```python
valores_brutos = [10, 50, 150, 20, 300]

def verificar_sistema(valores: list[int]) -> list[int]:
    alertas = []

    for valor in valores:
        if valor > 100:
            alertas.append(valor)
    return alertas

alertas_gerados = verificar_sistema(valores_brutos)

if len (alertas_gerados) == 0:
    print('Sistema Seguro')
else:
    print(f'⚠️Revisar valores: {alertas_gerados}')
```
